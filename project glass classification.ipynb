```python
mean = train_data.mean(axis=0) 
train_data -= mean
std = train_data.std(axis=0) 
train_data /= std
test_data -= mean 
test_data /= std
```

```python
k=5
num_val_samples = len(train_data) // k 
num_epochs = 100
```

```python
num_epochs=200

for i in range(k): 
    print('processing fold #', i)
    val_data = train_data[i * num_val_samples: (i + 1) * num_val_samples] 
    val_target = train_label[i * num_val_samples: (i + 1) * num_val_samples]
    
    
```

```python
train_data[1]
```

```python
from keras import models
from keras import layers
```

```python
model = models.Sequential()
```

```python
model.add(layers.Dense(20, activation='relu',input_shape=(train_data.shape[1],)))
model.add(layers.Dense(20, activation='relu'))
model.add(layers.Dense(1,))
```

```python
model.compile(optimizer='rmsprop', loss='mse', metrics=['acc'])
```

```python
history = model.fit(val_data,val_target,epochs=30,batch_size=10)
```

```python
a = model.evaluate(val_data,val_target)
```
